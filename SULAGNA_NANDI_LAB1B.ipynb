<h3> Introduction to the Lab </h3>

In mathematics, certain two-dimensional curves are best represented algebraically by expressing points $(x,y)$ on the curve in terms of a parameter $t$. We call these equations for $x$ and $y$, parametric equations, and we call the curve that is traced out by $(x,y)$ as $t$ varies a parametric curve. 

We often interpret this parametric curve $C$ as the curve traced out by a moving particle whose coordinates at time $t$ are given by $(x(t), y(t))$. For this reason, it makes sense to study trajectories, such as those seen in planetary motion, using parametric equations. When working with cyclical curves, like a planetary orbit, polar coordinates often prove to be more convenient than rectangular coordinates. 

In this lab, we will use code to explore curves described by parametric equations.  This will enable us to develop tools for studying planetary motion by integrating techniques of calculus and computer science.

<h3> Circles and Ellipses </h3>

The circle with centre $(h,k)$ and radius $r$ can be described by the parametric equations

$x = h+r\cos t$

$y = k+r\sin t$

where $0 \leq t \leq 2\pi$.


```python
## EXAMINE CODE IN THIS CELL, RUN IT TO SEE WHAT IT DOES

# parametric curve in R^2, part of unit circle 

# define parameter domain
n = 100
a = 0
b = 3 * np.pi / 2
t = np.linspace(a, b, n + 1)

# define parametric equations for curve C
x = np.cos(t)
y = np.sin(t)

# format plot
plt.plot(x, y)
plt.scatter(x[0], y[0], color="g", label="initial point")
plt.scatter(x[n], y[n], color="r", label="terminal point")
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')  # so it looks nice and circular
plt.title("Parametric Curve in $R^2$")
plt.grid()
plt.legend()
plt.show()
```

The position of one particle at time $t$ is given by 

$x=3\sin t \ \ \ \ \  y=2\cos t  \ \ \ \ \ 0 \leq t \leq 2\pi$

and the position of a second particle is given by 

$x=f(t) \ \ \ \ \  y=g(t)  \ \ \ \ \ 0 \leq t \leq 2\pi$


(a) Graph the paths of both particles. At how many points do the graphs intersect?

(b) Do the particles collide? If so, find the collision points.



```python
# code for problem 5
def intersections():
    t = np.linspace(0, 2 * np.pi, 40000)

    x1 = 3 * np.sin(t)
    y1 = 2 * np.cos(t)
    x2 = -3 + np.cos(t)
    y2 = -1 + np.sin(t)

    plt.grid()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim(-5, 5)  # define limits for x-axis
    plt.ylim(-2.5, 2.5)  # define limits for y-axis
    plt.plot(x1, y1)
    plt.plot(x2, y2)
    plt.show()

    distances = np.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    print(distances[distances < 0.0001])
    print(np.where(distances < 0.0001))
    print("Time of collision = " + str(t[np.where(distances < 0.0001)]))


intersections()
```

<h3> Parametric Curves in Higher Dimensions </h3>

By defining parametric equations for $x$, $y$, and $z$ in terms of $t$, we can represent space curves, i.e., curves in three dimensions.


```python
## EXAMINE CODE IN THIS CELL, RUN IT TO SEE WHAT IT DOES

# parametric curves in R^3

# define parameter domain
n = 100
a = 0
b = 20
t = np.linspace(a, b, n + 1)

# define parametric equations for curve C
x = t
y = 1 / (1 + t ** 2)
z = t ** 2

# format plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
ax.scatter(x[0], y[0], z[0], color="g", label="initial point")
ax.scatter(x[n], y[n], z[n], color="r", label="terminal point")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.title("Parametric Curve in $R^3$")
plt.legend()
plt.show()
```
